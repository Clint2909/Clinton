```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes' ]]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
#max_features='log2',
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid={'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],'n_estimators':[50,100],'max_features':['sqrt','log2',None],'random_state':[0]}

grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_

print("The R_score value for best parameters {}: ".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
y_pred=grid.predict(independent)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(dependent,y_pred)
```
